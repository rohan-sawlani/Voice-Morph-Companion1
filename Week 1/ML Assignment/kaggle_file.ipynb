```python
import numpy as np
import pandas as pd
df = pd.read_csv("/kaggle/input/glass/glass.csv")
df
```

```python
X= df.loc[:,["RI","Na","Mg","Al","Si","K","Ca","Ba","Fe"]]
y = df.loc[:,"Type"]
X=np.array(X)
y=np.array(y)
Xtrain = X[:180]
Xtest=X[180:]
ytrain = y[:180]
ytest = y[180:]
#3.part done
from collections import Counter
def ed(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
class KNN(object):
    def __init__(self,k):
        self.k=k
    def fit(self,Xtrain,ytrain):
        self.Xtrain=Xtrain
        self.ytrain=ytrain
    def predict(self,Xtest):
        predictions=[self._helper(x) for x in Xtest]
        return np.array(predictions)
    def _helper(self,x):
        prediction=[ed(x,x1) for x1 in self.Xtrain]
        indices= np.argsort(prediction)[:self.k]
        labels= [self.ytrain[i] for i in indices]
        c=Counter(labels).most_common()
        return c[0][0]
    
    
clf=KNN(k=3)
clf.fit(Xtrain,ytrain)
pred=clf.predict(Xtest)
pred
```

```python
def accuracy(predictions,y_test):
    return np.sum(predictions==y_test)/len(y_test)
accuracy(pred,ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf1= KNeighborsClassifier(3)
clf1.fit(Xtrain,ytrain)
pred1= clf1.predict(Xtest)
pred1
accuracy(pred1,ytest)
#both have same accuracy
```

###Question 1:
Generate a dataset for linear regression with 1000 samples, 5 features and single target.

Visualize the data by plotting the target column against each feature column. Also plot the best fit line in each case.

Hint : search for obtaining regression line using numpy.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
X, y = make_regression(n_samples=1000, n_features=5, noise=0 )  
fig, axes= plt.subplots(1,5,figsize=(20,4))
X0=X[:,np.newaxis,0]

X_train, X_test, y_train, y_test = train_test_split(X0, y, train_size = 0.8)

model=linear_model.LinearRegression()
model.fit(X_train,y_train)
ypred= model.predict(X_test)
axes[0].scatter(X_test,y_test)
axes[0].plot(X_test, ypred)

X1=X[:,np.newaxis,1]

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y, train_size = 0.8)

model=linear_model.LinearRegression()
model.fit(X1_train,y1_train)
ypred1= model.predict(X1_test)
axes[1].scatter(X1_test,y1_test)
axes[1].plot(X1_test, ypred1)

X2=X[:,np.newaxis,2]

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, train_size = 0.8)

model=linear_model.LinearRegression()
model.fit(X2_train,y2_train)
ypred2= model.predict(X2_test)
axes[2].scatter(X2_test,y2_test)
axes[2].plot(X2_test, ypred2)
X3=X[:,np.newaxis,3]

X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y, train_size = 0.8)

model=linear_model.LinearRegression()
model.fit(X3_train,y3_train)
ypred3= model.predict(X3_test)
axes[3].scatter(X3_test,y3_test)
axes[3].plot(X3_test, ypred3)

X4=X[:,np.newaxis,4]

X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y, train_size = 0.8)

model=linear_model.LinearRegression()
model.fit(X4_train,y4_train)
ypred4= model.predict(X4_test)
axes[4].scatter(X4_test,y4_test)
axes[4].plot(X4_test, ypred4)
fig
plt.tight_layout()



```

## Question 4
Go to the website https://www.worldometers.info/coronavirus/ and scrape the table containing covid-19 infection and deaths data using requests and BeautifulSoup.  Convert the table to a Pandas dataframe with the following columns : Country, Continent, Population, TotalCases, NewCases, TotalDeaths, NewDeaths,TotalRecovered, NewRecovered,  ActiveCases.

*(<b>Optional Challenge :</b> Change the data type of the Columns (Population ... till ActiveCases) to integer. For that you need to remove the commas and plus signs. You may need to use df.apply() and pd.to_numeric() . Take care of the values which are empty strings.)

```python
import requests as req
import pandas as pd
url = 'https://www.worldometers.info/coronavirus/'
page= req.get(url)
htmlContent = page.content
from bs4 import BeautifulSoup
soup = BeautifulSoup(htmlContent, 'html.parser')
table = soup.find(id="main_table_countries_today")
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

rows=[]
for tr in table.find_all('tr'):
    cells = tr.find_all('td')
    row=[cell.text.strip() for cell in cells]
    rows.append(row)
df = pd.DataFrame(rows,columns=headers)
df = df[['Country,Other', 'Continent', 'Population', 'TotalCases', 'NewCases', 'TotalDeaths', 'NewDeaths', 'TotalRecovered', 'NewRecovered', 'ActiveCases']]
df.columns = ['Country', 'Continent', 'Population', 'TotalCases', 'NewCases', 'TotalDeaths', 'NewDeaths', 'TotalRecovered', 'NewRecovered', 'ActiveCases']

df[1:]
#done
```

# Question 5

Generate an imbalanced classification dataset using sklearn of 1000 samples with 2 features, 2 classes and 1 cluster per class. Plot the data. One of the class should contain only 5% of the total samples. Confirm this either using numpy or Counter. Plot the data.

Now oversample the minority class to 5 times its initial size using SMOTE. Verify the number. Plot the data.

Now undersample the majority class to 3 times the size of minority class using RandomUnderSampler. Verify the number. Plot the data.

Reference : Last markdown cell of the examples.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from collections import Counter

# Step 1: Generate the imbalanced dataset
X, y = make_classification(n_samples=1000, 
                           n_features=2, 
                           n_informative=2, 
                           n_redundant=0, 
                           n_clusters_per_class=1, 
                           weights=[0.95, 0.05], 
                           flip_y=0, 
                           random_state=42)

# Step 2: Confirm the class distribution
class_distribution = Counter(y)
print("Class distribution:", class_distribution)
plt.figure(figsize=(8,6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1])
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1])

     
```

```python
import seaborn as sns
from sklearn.datasets import make_blobs
from imblearn.over_sampling import SMOTE
Xdf = pd.DataFrame(X)
ydf= pd.DataFrame(y)
Xdf = pd.concat([
    Xdf[y == 0],
    Xdf[y == 1].sample(40, random_state=42)
], axis=0)
ydf = ydf.loc[Xdf.index]
sm = SMOTE(
    
    sampling_strategy=0.2642,
    random_state=0,
    k_neighbors=5,
)
Xdf_res, ydf_res = sm.fit_resample(Xdf, ydf)
Xdf_res.shape, ydf_res.shape
```

```python
plt.figure(figsize=(8,6))
X1 =Xdf_res.to_numpy()
y1= ydf_res.to_numpy().transpose().reshape(1200)
plt.scatter(X1[y1 == 0][:, 0], X1[y1 == 0][:, 1])
plt.scatter(X1[y1 == 1][:, 0], X1[y1 == 1][:, 1])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(
    sampling_strategy=0.250,
    random_state=0, 
)  
Xdf_resampled, ydf_resampled = rus.fit_resample(Xdf, ydf)
Xdf_resampled.shape
```

```python
plt.figure(figsize=(8,6))
X1 =Xdf_resampled.to_numpy()
y1= ydf_resampled.to_numpy().transpose().reshape(200)
plt.scatter(X1[y1 == 0][:, 0], X1[y1 == 0][:, 1])
plt.scatter(X1[y1 == 1][:, 0], X1[y1 == 1][:, 1])
#done
```
